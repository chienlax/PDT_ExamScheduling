```python
import pandas as pd
```

```python
# Đọc file CSV vào DataFrame
df = pd.read_csv('HCM.csv', encoding='utf-8-sig')

column_name = df.columns[4]  # lấy tên cột đầu tiên

# Drop cột
df = df[df[column_name] != 'TXGDTC0001']
df = df[df[column_name] != 'TXGDTC0002']
df = df[df[column_name] != 'TXGDTC0003']
df = df[df[column_name] != 'TXGDTC0004']

# Sắp xếp DataFrame theo cột đầu tiên
df = df.sort_values(by=df.columns[4])
df.to_csv("HCM_sorted.csv", encoding='utf-8-sig', index=False)
```

```python
# Chia batch nhỏ
df1 = pd.read_csv('HCM_sorted.csv', encoding='utf-8-sig')

first_346_rows = df1.head(346)
# Ghi DataFrame vào file CSV mới
first_346_rows.to_csv('HCM_346rows.csv', index=False, encoding='utf-8-sig')

selected_rows_347_653 = pd.concat([df.head(1), df.iloc[347:653]])

# Export the selected rows to a new CSV file
selected_rows_347_653.to_csv('HCM_347rows.csv', index=False)
```

```python
df3 = pd.read_csv('HCM_0-346.csv', encoding='utf-8-sig')
df3 = df3.sort_values(by=df.columns[1])
df3.to_csv("HCM_0-346_sortedMSV.csv", encoding='utf-8-sig', index=False)
```

```python
df4 = pd.read_csv('HCM_0-346_sortedMSV.csv', encoding='utf-8-sig')
d = dict()
for value in df4[df.columns[1]]:
    if value not in d:
        d[value] = 1
    else:
        d[value]+=1

cnt = 0
for value in d:
    if d[value] > 3:
        cnt += 1
        print(value)
print(d)
print(cnt)
```
